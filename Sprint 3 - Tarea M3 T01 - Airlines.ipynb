```python
# Meses con mayor retraso : 

MonthDelay = df.groupby(['Month'])[['Month']].agg(['count'])
MonthDelay.sort_values(by=[('Month', 'count')], ascending=False)
```

```python
# Días de la semana con mayor retraso : 

DayDelay = df.groupby(['DayOfWeek'])[['DayOfWeek']].agg(['count'])
DayDelay.sort_values(by=[('DayOfWeek', 'count')], ascending=False)
```

```python
# Origen con mayor DepDelay acumulado:

OriginDepDelay = df.groupby(['Origin'])[['DepDelay']].agg(['sum'])
OriginDepDelay = OriginDepDelay.sort_values(by=[('DepDelay', 'sum')], ascending=False)
OriginDepDelay.head(10)
```

```python
# Destino con mayor ArrDelay acumulado: 

DestDepDelay = df.groupby(['Dest'])[['ArrDelay']].agg(['sum'])
DestDepDelay = DestDepDelay.sort_values(by=[('ArrDelay', 'sum')], ascending=False)
DestDepDelay.head(10)
```

```python
# Vuelos más largos:

MoreDist = df.groupby(['Origin', 'Dest'])[['Distance']].agg(['max'])
MoreDist = MoreDist.sort_values(by=[('Distance', 'max')], ascending=False)
MoreDist.head(10)
```

```python
#Crear nuevas columnas:

    # Velocidad media de vuelo/ Average Flight Speed (AveFliSpeed) expresado en miles/h:
df["AveFliSpeed"] = round((df["Distance"] / df["ActualElapsedTime"])*60,2)
```

```python
df.head()
```

```python
# Ha llegado tarde o no (Columna On Time):

import numpy as np

df['OnTime'] = np.where(df['ArrTime'] <= df['CRSArrTime'], True, False)
df.head()
```

```python
# Que tan tarde a llegado (Range of Delay):

# create a list of our conditions
conditions = [
    (df['ArrDelay'] <= 2),
    (df['ArrDelay'] > 2) & (df['ArrDelay'] <= 7),
    (df['ArrDelay'] > 7) & (df['ArrDelay'] <= 15),
    (df['ArrDelay'] > 15)
    ]

# create a list of the values we want to assign for each condition
values = ['No_Delay', 'Min_Delay', 'Modarate_Delay', 'Very_Delay']

# create a new column and use np.select to assign values to it using our lists as arguments
df['RangeDelay'] = np.select(conditions, values)

# display updated DataFrame
df.head()
```

```python
# Aerolíneas con mayores retrasos acumulados:
# A pesar de salir antes o después, el retreaso real esta a la hora de llegada

AcuArrDelay = df.groupby(['UniqueCarrier'])[['ArrDelay']].agg(['sum'])
AcuArrDelay = AcuArrDelay.sort_values(by=[('ArrDelay', 'sum')], ascending=False)
AcuArrDelay.head()
```

```python
# Similar al paso anterior, pero SÓLO considerando los retrasos de llegada (y NO los tiempos de llegada antes de lo previsto)
# Creo una nueva columna para realizar el calculo

conditions = [
    (df['ArrDelay'] <= 0), #Valores negativos los ponemos en cero para que no alteren los retrasos 
    (df['ArrDelay'] > 0)
    ]

# create a list of the values we want to assign for each condition
values = [0, df['ArrDelay']] 

# create a new column and use np.select to assign values to it using our lists as arguments
df['RealDelay'] = np.select(conditions, values)

df.head()
```

```python
# Aerolíneas con mayores retrasos acumulados:
# display updated DataFrame

AcuArrDelay2 = df.groupby(['UniqueCarrier'])[['RealDelay']].agg(['sum'])
AcuArrDelay2 = AcuArrDelay2.sort_values(by=[('RealDelay', 'sum')], ascending=False)
AcuArrDelay2.head()
```

```python
# Rutas más largas y las que acumulen más retrasos:

MoreDist2 = df.groupby(['Distance','Origin', 'Dest'])[['RealDelay']].agg(['sum'])
MoreDist2.sort_index(ascending = False).head(15)
```

```python
# Vuelos con mayor Elapsed Time que el programado:

conditions = [
    ((df["ActualElapsedTime"] - df["CRSElapsedTime"]) <= 0),  
    ((df["ActualElapsedTime"] - df["CRSElapsedTime"]) > 0)
    ]

# create a list of the values we want to assign for each condition
values = [0, (df["ActualElapsedTime"] - df["CRSElapsedTime"])] 

# create a new column and use np.select to assign values to it using our lists as arguments
df['ExtraET'] = np.select(conditions, values)

df.head()
```

```python
# Vuelos con Mayor Extra Elapsed Time (Tiempo acumulado y numero de casos):

ExtraETFlights = df.groupby(['Distance','Origin', 'Dest'])[['ExtraET']].agg(['sum', 'count' ])
ExtraETFlights = ExtraETFlights.sort_values(by=[('ExtraET', 'sum')], ascending=False)
ExtraETFlights.head()
```

```python
# Otros motivos de retraso según ruta (Tiempo acumulado y tiempo máximo):

OthersDelays = df.groupby(['Origin', 'Dest'])[['Diverted',
    'CarrierDelay', 'WeatherDelay', 'NASDelay', 
    'SecurityDelay', 'LateAircraftDelay']].agg(['sum', 'count', 'max' ])
OthersDelays.head(10)
```
